```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 
%matplotlib notebook

df = pd.read_csv('acc_dat.csv')
df = df[df['STATE/UT'] == 'Delhi (Ut)']
df.drop(['STATE/UT', 'Total'], axis = 1, inplace = True)
df = df.iloc[9:]
df.columns = ['Year', '3AM', '6AM', '9AM', '12PM', '3PM', '6PM', '9PM', '12AM']

df.set_index('Year',inplace = True)

df = df.T

df.index.names = ['Timing']

df
```

```python
df.plot.line()
plt.ylabel('Number of Accidents')
plt.title('Frequency of Accidents during different parts of the day in New Delhi')
```
